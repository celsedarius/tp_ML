## 3. entrainement des modeles

### preparation des donnees

```python
# on efface la colonne ID puis que n'etant pas utile
df = df.drop('ID')
df.dtypes
```

#### toutes les donnees sont en int, donc on aura pas besoin de transformer les donnnes categorie

```python
#colonne a vectoriser
col = [col for col in df.columns if col!= 'DEFAULT'  ]
col
```

```python
for c in col:
    count=df.filter(c+" is   NULL" or c+"is ''" or c+"is  NaN" or c+"is  null").count()
    print(c+" column has "+ str(count) +" missing values\n")
```

Aucune valeur manquante donc pas besoin d'Imputer pour completer les donnees

### scaling

```python
from pyspark.ml.feature import VectorAssembler, StandardScaler


va =  VectorAssembler().setInputCols(col).setOutputCol("features_to_be_scale")

df_feature = va.transform(df)

scaler =  StandardScaler().setInputCol("features_to_be_scale").setOutputCol("features")
scaled_df = scaler.fit(df_feature).transform(df_feature).select("features", "DEFAULT")

scaled_df.show(5)
```

### division du dataset

```python
trainSet, testSet = scaled_df.randomSplit([0.7,0.3])
```

### a. Logistic Regression

```python
from pyspark.ml.classification import LogisticRegression

lr= LogisticRegression(labelCol="DEFAULT")

logisticModel= lr.fit(trainSet)
lr_prediction = logisticModel.transform(testSet)
log_summary=logisticModel.summary
log_summary.predictions.show()
```

### b. DecisionTree

```python
from pyspark.ml.classification import DecisionTreeClassifier


dt = DecisionTreeClassifier(featuresCol = 'features', labelCol = 'DEFAULT')

dtModel = dt.fit(trainSet)
dt_predictions = dtModel.transform(testSet)
dt_predictions.show(10)
```

### c.Random Forest

```python
from pyspark.ml.classification import RandomForestClassifier
rf = RandomForestClassifier(featuresCol = 'features', labelCol = 'DEFAULT')
rfModel = rf.fit(trainSet)
rf_predictions = rfModel.transform(testSet)
rf_predictions.show(10)
```

### comparaison des performances

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator

evaluator= BinaryClassificationEvaluator().setLabelCol("DEFAULT")

# print("logistic regression      Test error: {:.4f}".format(1-evaluator.evaluate(lr_prediction)))
# print("data tree regression     Test error: {:.4f}".format(1-dt_evaluator.evaluate(dt_prediction)))
# print("random forest regression Test error: {:.4f}".format(1-rf_evaluator.evaluate(rf_prediction)))


lr_roc = evaluator.evaluate(lr_prediction, {evaluator.metricName: "areaUnderROC"})
dt_roc = evaluator.evaluate(dt_predictions, {evaluator.metricName: "areaUnderROC"})
rf_roc = evaluator.evaluate(rf_predictions, {evaluator.metricName: "areaUnderROC"})


print("Logistic regression      Area under ROC Curve: {:.4f}".format(lr_roc))
print("data tree regression     Area under ROC Curve: {:.4f}".format(dt_roc))
print("random forest regression Area under ROC Curve: {:.4f}".format(rf_roc))
```

## 4. Amelioration du resultat

pour ameliorer le resultat, nous allons utiliser le meilleur modele ici, et faire de l"`hyperparameter tuning`"
ceci permettra de choisir les parametres optimal afin d'ameliorer le resultat

```python
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
paramGrid = (ParamGridBuilder()
             .addGrid(rf.maxDepth, [5,10])
             .addGrid(rf.numTrees, [ 5, 10])
             .addGrid(rf.maxBins, [20, 60])            
             .build())

cv = CrossValidator(estimator=rf, estimatorParamMaps=paramGrid, evaluator=evaluator, numFolds=5)

cvModel = cv.fit(trainSet)
predictions = cvModel.transform(testSet)
evaluator.evaluate(predictions)
```
